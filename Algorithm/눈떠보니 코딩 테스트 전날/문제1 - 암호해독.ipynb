```
text = ['   + -- + - + -   ',
        '   + --- + - +   ',
        '   + -- + - + -   ',
        '   + - + - + - +   ']
```

```
for i in text:
  print(i)

# lstrip : 앞 공백 제거 , rstrip : 뒤 공백 제거, strip : 앞 뒤 둘다 제거

for i in text:
  print(i.strip())
```

```
for i in text:
    print(i.strip().replace(' ', '').replace('+', '1').replace('-', '0'))
```

```
for i in text:
    print(int(i.strip().replace(' ', '').replace('+', '1').replace('-', '0'))) # 10진법 변환    
```

```
for i in text:
    print(int(i.strip().replace(' ', '').replace('+', '1').replace('-', '0'), 2)) # 10진법 변환, 문자열을 2진법으로 인식해서 10진수로 바꿔준다.    
```

```
# ord() : 문자 => 숫자
# chr() : 숫자 => 문자

l = []
for i in text:
    l.append(chr(int(i.strip().replace(' ', '').replace('+', '1').replace('-', '0'), 2)))
    print(chr(int(i.strip().replace(' ', '').replace('+', '1').replace('-', '0'), 2))) # 10진법 변환, 문자열을 2진법으로 인식해서 10진수로 바꿔준다.    

print(l)   


ord_list = list(l)

for i in ord_list:
  print(ord(i))
  
```

```
# map(function, 순회 객체)

print(map(lambda x : chr(int(x, 2)), s)) # 주소값만 나옴.

list(map(lambda x : chr(int(x, 2)), s))
```
